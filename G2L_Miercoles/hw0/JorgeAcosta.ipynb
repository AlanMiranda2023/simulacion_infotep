```python
# prompt: Cree las siguiente funciones a la clase escena:
# crear_escena: Crea una instancia de una escena que deberá contener una circunferencia de tamaño y de un color dado por parametro. Tambien contendrá 1 punto superpuesto a la circunferencia.
# cambiar_escena: Se podrá cambiar los parametros del objeto escena
# dibujar_escena: Deberá generar una representación grafica de la escena
# desplazar_puntos: Esta función realizará un desplazamiento de valor x en una dirección definida por parametro (derecha, izquierda, arriba, abajo)
# simular: Esta función deberá recibir como parametro una velocidad y una dirección y se debe generar una animación del desplazamiento del punto

import matplotlib.pyplot as plt
import numpy as np

class Escena:
    def __init__(self):
        self.circulo_radio = 0
        self.circulo_color = 'black'
        self.punto_x = 0
        self.punto_y = 0
        self.fig, self.ax = plt.subplots()

    def crear_escena(self, radio, color):
        self.circulo_radio = radio
        self.circulo_color = color
        # Calcular la posición del punto superpuesto a la circunferencia
        self.punto_x = radio
        self.punto_y = radio

    def cambiar_escena(self, radio=None, color=None):
        if radio:
            self.circulo_radio = radio
        if color:
            self.circulo_color = color

    def dibujar_escena(self):
        # Dibujar la circunferencia
        circulo = plt.Circle((0, 0), self.circulo_radio, color=self.circulo_color, fill=False)
        self.ax.add_patch(circulo)
        # Dibujar el punto
        self.ax.scatter(self.punto_x, self.punto_y, color='black')
        # Configurar los límites del gráfico
        self.ax.set_xlim(-self.circulo_radio * 1.2, self.circulo_radio * 1.2)
        self.ax.set_ylim(-self.circulo_radio * 1.2, self.circulo_radio * 1.2)
        # Mostrar el gráfico
        plt.show()


    def desplazar_puntos(self, valor_x, direccion):
        if direccion == "derecha":
            self.punto_x += valor_x
        elif direccion == "izquierda":
            self.punto_x -= valor_x
        elif direccion == "arriba":
            self.punto_y += valor_x
        elif direccion == "abajo":
            self.punto_y -= valor_x

    def simular(self, velocidad, direccion):
        # Implementación básica de la animación (se puede mejorar con FuncAnimation)
        for _ in range(10):  # Simular por 10 pasos
            self.desplazar_puntos(velocidad, direccion)
            self.dibujar_escena()
            # plt.pause(0.1)  # Pausar brevemente para visualizar el cambio
```

```python
# prompt: crear un ejemplo de uso de la clase anterior

# Ejemplo de uso de la clase Escena

# Crear una instancia de la clase Escena
escena = Escena()

# Crear la escena inicial con un círculo de radio 5 y color azul
escena.crear_escena(10, 'blue')

# Dibujar la escena inicial
escena.dibujar_escena()

# Cambiar el color del círculo a rojo
escena.cambiar_escena(color='black')

# Dibujar la escena con el nuevo color
escena.dibujar_escena()

# Desplazar el punto 2 unidades a la derecha
escena.desplazar_puntos(6, "izquierda")
escena.dibujar_escena()

# Simular el movimiento del punto 1 unidad a la izquierda durante 10 pasos
escena.simular(1, "derecha")


#Cambiar el radio del circulo a 10
escena.cambiar_escena(radio=10)
escena.dibujar_escena()

```
