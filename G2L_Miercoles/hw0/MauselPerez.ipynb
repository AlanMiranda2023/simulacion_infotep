```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

#MAUSEL PEREZ

class Scene:
  def __init__(self, size, colour, position):
    self.size = size
    self.colour = colour
    self.position = position
    self.fig, self.ax = plt.subplots()
    self.circle = plt.Circle((0, 0), self.size, color=self.colour, fill=False)
    self.ax.add_artist(self.circle)
    self.ax.plot(self.position[0], self.position[1], 'ro')
    self.ax.set_xlim(-self.size-1, self.size+1)
    self.ax.set_ylim(-self.size-1, self.size+1)
    self.ax.set_aspect('equal', 'box')
    plt.title("TEST DE ESCENA 2D")

  #Crear escena
  def create_scene(self, size, colour, position):
     self.size = size
     self.colour = colour
     self.position = position
     self.circle = plt.Circle((0, 0), self.size, color=self.colour, fill=False)
     self.ax.clear()
     self.ax.add_artist(self.circle)
     self.ax.plot(self.position[0], self.position[1], 'ro')
     self.ax.set_xlim(-self.size-1, self.size+1)
     self.ax.set_ylim(-self.size-1, self.size+1)
     self.ax.set_aspect('equal', 'box')
     plt.title("TEST DE ESCENA 2D")
     plt.draw()

  #Cambiar de escena
  def change_scene(self, size, colour, position):
    self.create_scene(size, colour, position)

  #Mostrar escena
  def draw_scene(self):
    plt.show()

  #Mover puntos
  def move_point(self, direction, distance):
    if direction == "derecha":
      self.position = (self.position[0] + distance, self.position[1])

    elif direction == "izquierda":
      self.position = (self.position[0] - distance, self.position[1])

    elif direction == "arriba":
      self.position = (self.position[0], self.position[1] + distance)

    elif direction == "abajo":
      self.position = (self.position[0], self.position[1] - distance)

    self.ax.clear()
    self.create_scene(self.size, self.colour, self.position)


  def simulate(self, speed, direction, time=5):
        num_frames = int(time * speed)

        def update(frame):
            if direction == "derecha":
                new_position = (self.position[0] + frame, self.position[1])
            elif direction == "izquierda":
                new_position = (self.position[0] - frame, self.position[1])
            elif direction == "arriba":
                new_position = (self.position[0], self.position[1] + frame)
            elif direction == "abajo":
                new_position = (self.position[0], self.position[1] - frame)

            self.ax.clear()
            self.create_scene(self.size, self.colour, new_position)

        ani = animation.FuncAnimation(self.fig, update, frames=range(0, num_frames), interval=100, repeat=False)
        plt.show()


scene = Scene(10, 'blue', (2, 3))

scene.change_scene(12, 'blue', (5, 6))

scene.move_point("abajo", 10)



```
