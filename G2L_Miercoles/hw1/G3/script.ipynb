```python
import matplotlib.pyplot as plt
import time

def crear_escena(radio, color_circulo, posicion_punto):
    return {
        "radio": radio,
        "color_circulo": color_circulo,
        "punto_x": posicion_punto[0],
        "punto_y": posicion_punto[1]
    }

def cambiar_escena(escena, nuevo_radio, nuevo_color, nueva_posicion):
    escena["radio"] = nuevo_radio
    escena["color_circulo"] = nuevo_color
    escena["punto_x"], escena["punto_y"] = nueva_posicion

def dibujar_escena(escena):
    fig, ax = plt.subplots()
    circulo = plt.Circle((0, 0), escena["radio"], color=escena["color_circulo"], alpha=0.5)
    ax.add_patch(circulo)
    ax.plot(escena["punto_x"], escena["punto_y"], 'ro', markersize=8)
    ax.set_xlim(-escena["radio"]-1, escena["radio"]+1)
    ax.set_ylim(-escena["radio"]-1, escena["radio"]+1)
    ax.set_aspect('equal')
    plt.grid()
    plt.show()

def desplazar_punto(escena, desplazamiento, direccion):
    if direccion == "derecha":
        escena["punto_x"] += desplazamiento
    elif direccion == "izquierda":
        escena["punto_x"] -= desplazamiento
    elif direccion == "arriba":
        escena["punto_y"] += desplazamiento
    elif direccion == "abajo":
        escena["punto_y"] -= desplazamiento
    else:
        print("Dirección no válida")

def simular(escena, velocidad, direccion, pasos=10):
    for _ in range(pasos):
        desplazar_punto(escena, 1, direccion)
        dibujar_escena(escena)
        time.sleep(velocidad)

escena = crear_escena(radio=5, color_circulo='blue', posicion_punto=(0, 0))
dibujar_escena(escena)
simular(escena, 0.5, "derecha", 5)
```
