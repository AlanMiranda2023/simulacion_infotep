```python
import matplotlib.pyplot as plt
import numpy as np

class Escena:
    def __init__(self, size_circumference=1, color_circumference='blue', punto=(0,0)):
        self.size_circumference = size_circumference
        self.color_circumference = color_circumference
        self.punto = np.array(punto, dtype=np.float64)

    def crear_escena(self):
        print(f"Escena creada con circunferencia de tamaño {self.size_circumference} y color {self.color_circumference}")
        print(f"Punto inicial en coordenadas: {self.punto}")


    def cambiar_escena(self, new_size=None, new_color=None, new_punto=None):
        if new_size:
            self.size_circumference = new_size
        if new_color:
            self.color_circumference = new_color
        if new_punto:
            self.punto = np.array(new_punto)
        print("Parámetros de la escena modificados.")


    def dibujar_escena(self):
        circle = plt.Circle((0, 0), self.size_circumference, color=self.color_circumference, fill=False)
        fig, ax = plt.subplots()
        ax.add_patch(circle)
        plt.scatter(self.punto[0], self.punto[1], color='red')
        ax.set_aspect('equal')
        plt.xlim(-self.size_circumference*2, self.size_circumference*2)
        plt.ylim(-self.size_circumference*2, self.size_circumference*2)
        plt.title("Escena 2D")
        plt.show()

    def desplazar_puntos(self, valor_desplazamiento, direccion):
        direcciones = {
            "derecha":  np.array([valor_desplazamiento, 0]),
            "izquierda": np.array([-valor_desplazamiento, 0]),
            "arriba":   np.array([0, valor_desplazamiento]),
            "abajo":    np.array([0, -valor_desplazamiento])
        }

        if direccion in direcciones:
            self.punto += direcciones[direccion]
        else:
            print("Dirección no válida.")

    def simular(self, velocidad, direccion, pasos=10):
        for _ in range(pasos):
            self.desplazar_puntos(velocidad, direccion)
            self.dibujar_escena()


escena_2D = Escena(size_circumference=7, punto=(4, 4))
escena_2D.crear_escena()
escena_2D.dibujar_escena()
escena_2D.desplazar_puntos(0.7, 'derecha')
escena_2D.dibujar_escena()
escena_2D.simular(0.2, 'arriba', pasos=5)
```
